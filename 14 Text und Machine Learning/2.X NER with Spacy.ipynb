## Mit Spacy NLP verarbeiten Named Entity= Eigennamenerkennung

```python
article = nlp(article) #Hier zeigt er an, wieviele NLP dass es hat
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels) #von welcher Art Eigennamen hats wieviele
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in article.sents] #wir splitten einen Text auf
#print(sentences[2])
sentences[3]
```

```python
displacy.render(nlp(str(sentences[4])), jupyter=True, style='ent') 
#Wir nehmen einen SAtz und lassen diesen verarbeiten
```

```python
displacy.render(nlp(str(sentences[0])), style='dep', jupyter = True, options = {'distance': 120})
```

```python
[(x.orth_,x.pos_, x.lemma_) for x in [y 
                                      for y
                                      in nlp(str(sentences[0])) 
                                      if not y.is_stop and y.pos_ != 'PUNCT']]
```

```python
dict([(str(x), x.label_) for x in nlp(str(sentences[4])).ents])
```

```python
displacy.render(article, jupyter=True, style='ent')
```
