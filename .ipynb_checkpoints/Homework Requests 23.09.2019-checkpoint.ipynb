2. Mache das gleiche für August 2017, 2016, 2015?

```python
#meine Lösung. 
liste_beben=[]
for year in range(2015,2019):
    r= requests.get("https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=" +str(year)+"-08-01&endtime="+ str(year)+"-08-31")
  
    liste_beben.append(r.text)
print(liste_beben)
    
#print umbedingt wie hier gemacht wieder aus dem Loop nehmen, sonst druckt er jedesmal aus.
```

```python
#Retos lösung ums zusammenzuziehen
"https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2014-01-01&endtime=2014-01-02"

base = "https://earthquake.usgs.gov/fdsnws/event/1/"
action = "count?"
# Url = "starttime=2018-09-25&endtime=2018-09-30-25&endtime=2018-09-30"

erdbebenanzahl=[]

for jahr in range(2015, 2019):
    
    url = base + action + "starttime=" + str(jahr) + "-08-01&endtime=" + str(jahr) + "-08-31"
    
    response = requests.get(url)

    print(str(jahr) + ": " + str(response.text))
    
    minidict={"Jahr":jahr, "Anzahl":str(response.text)}
    erdbebenanzahl.append(minidict)
```

```python
pd.DataFrame(erdbebenanzahl)
```

5. Wo haben sich diese Erdbeben ereignet?

### Lösung von Jörg

```python
erdbeben_max = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-01-01&endtime=2019-09-24&minmagnitude=6.5")
erdbeben_max = erdbeben_max.json()
erdbeben_max = erdbeben_max["features"] #er schaut an, was alles drin ist

#nun schauen wir die SAchen an
for beben in erdbeben_max:
    print(beben["properties"]["mag"])
    print(beben["properties"]["place"])
    print(beben["properties"]["time"])
    
import time 
top2019 = []
#nun gehts an das rausziehen und ins dict versorgen.
for beben in erdbeben_max:
    magnitude = beben["properties"]["mag"]
    place = beben["properties"]["place"]
    zeit = beben["properties"]["time"]

    mini_dict = {"Magnitude: ":magnitude, "Place: ":place, "Time: ":zeit}
    
    top2019.append(mini_dict)
```

```python
pd.DataFrame(top2019).sort_values("Magnitude: ", ascending=False)[:10]
```

6. Nun noch die Zeit richtig darstellen - so dass es nicht in unix ist

```python
#die Zeit ist hier in unix mit Millisekunden. die braucht man nicht. deshalb die letzten drei abschneiden.
from datetime import datetime
import time
erdbeben_max = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-01-01&endtime=2019-09-24&minmagnitude=6.5")
erdbeben_max = erdbeben_max.json()
erdbeben_max = erdbeben_max["features"] 

top2019 = []
for beben in erdbeben_max:
    magnitude = beben["properties"]["mag"]
    place = beben["properties"]["place"]
    
    zeit = str(beben["properties"]["time"])[:-3] #hier ändere ich nun was: ich schneide die letzten drei Zahlen ab
    zeit= datetime.utcfromtimestamp(int(zeit))
    zeit= zeit.strftime("%Y-%m-%d %H:%M:%S")
    #unter strftime.org findet man alle entsprechenden Abkürzungen für die Zeitangabe.
    

    mini_dict = {"Magnitude: ":magnitude, "Place: ":place, "Time: ":zeit}
    
    top2019.append(mini_dict)
```

```python
pd.DataFrame(top2019).sort_values("Magnitude: ", ascending=False)[:10]
```
